```
import os
import random
from shutil import copy2, make_archive
from distutils.dir_util import copy_tree
```

```
# Root folder. Change this according to your folder structure.
ROOT_DIR = '/deeplearning/'
```

```
# Get original zipped data.
zipped_data = os.path.join(ROOT_DIR, 'cse7512-00-dl-tp-pt1.zip')

data_dir = '/content/data'
os.makedirs(data_dir, exist_ok=True)
!cp '{zipped_data}' '{data_dir}'

train_dir = os.path.join(data_dir, 'train')
test_dir = os.path.join(data_dir, 'test')
```

```
# Unzip the data.
!unzip -qq '{data_dir}/cse7512-00-dl-tp-pt1.zip' -d '{data_dir}'
```

```
# Create necessary folders.
processed_data_dir = '/content/processed_data'
os.makedirs(processed_data_dir, exist_ok=True)

processed_train_dir = os.path.join(processed_data_dir, 'train')
processed_validation_dir = os.path.join(processed_data_dir, 'validation')

os.makedirs(processed_train_dir, exist_ok=True)
os.makedirs(processed_validation_dir, exist_ok=True)
```

```
# Divide the original train data into train and validation data.
# (400 train images and 100 validation images for each class.)
for class_label in os.listdir(train_dir):

    print('Label: ', class_label)
    train_class_dir = os.path.join(processed_train_dir, class_label)
    os.makedirs(train_class_dir, exist_ok=True)

    validation_class_dir = os.path.join(processed_validation_dir, class_label)
    os.makedirs(validation_class_dir, exist_ok=True)

    img_list = os.listdir(os.path.join(train_dir, class_label))
    print('Number of images: ', len(img_list))

    random.shuffle(img_list)
    for img in img_list[:400]:
        
        src = os.path.join(train_dir, class_label, img)
        dst = train_class_dir
        copy2(src, dst)
    print('Train: ', len(os.listdir(train_class_dir)))


    for img in img_list[400:]:
        
        src = os.path.join(train_dir, class_label, img)
        dst = validation_class_dir
        copy2(src, dst)
    print('Validation: ', len(os.listdir(validation_class_dir)), '\n')
```

```
# Copy test data.
!cp -R '{test_dir}' '{processed_data_dir}'
```

```
# Create a new zip file that contain train, validation and test floders.
make_archive('processed_data', 'zip', processed_data_dir)
```
